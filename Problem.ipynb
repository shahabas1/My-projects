# Algorithm

```python
import sklearn.model_selection
#Split into train and test
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x2, y2, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
m=RandomForestClassifier(n_estimators=1, max_depth=3, bootstrap=False, n_jobs=-1)
m.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=m.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
m1 = RandomForestClassifier(n_estimators=80, n_jobs=-1)
m1.fit(x_train, y_train)
y_pred1=m.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```
