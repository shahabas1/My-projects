```python
mean3=pd.to_numeric(df1[df1["TotalCharges"]!=' ']["TotalCharges"]).mean()
```

```python
df1['TotalCharges']=pd.to_numeric(df1['TotalCharges'].replace(' ',mean3))
```

```python
df1['TotalCharges']
```

```python
df1['churn1']=[1 if x == 'Yes' else 0 for x in df1['Churn']]
```

```python
df1['churn1']
```

```python
df1= df1.drop('Churn',axis=1)
```

```python
df1.head()
```

```python
df2= pd.get_dummies(df1)
df2.shape
```

```python
x=df2.drop('churn1',axis=1)
y=df2['churn1']
```

```python
y
```

```python
x.columns
```

```python
import sklearn.model_selection
#Split into train and test
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.25, random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
m = RandomForestClassifier(n_estimators=1, max_depth=3, bootstrap=False, n_jobs=-1)
m.fit(x_train, y_train)
y_pred=m.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
m = RandomForestClassifier(n_estimators=80, n_jobs=-1)
m.fit(x_train, y_train)
y_pred=m.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
